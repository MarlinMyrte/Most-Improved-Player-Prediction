```python
# Dropping duplicate rows
all_data=all_data.drop_duplicates()
```

```python
# Some players were traded mid-season. For them we want to keep only the Totals of that season
all_data["traded"]=0
for i in range(len(all_data)):
    if all_data.iloc[i,3]=="TOT":
        year=all_data.iloc[i,1]
        player=all_data.iloc[i,0]
        l=len(all_data[(all_data["br_id"]==player) & (all_data["season"]==year)])-1
        while l!=0:
            all_data.iloc[i+l,-1]+=1
            l=l-1
            
all_data=all_data[all_data["traded"]==0]

all_data.drop(columns=["traded"],inplace=True)

all_data_test["traded"]=0
for i in range(len(all_data_test)):
    if all_data_test.iloc[i,3]=="TOT":
        year=all_data_test.iloc[i,1]
        player=all_data_test.iloc[i,0]
        l=len(all_data_test[(all_data_test["br_id"]==player) & (all_data_test["season"]==year)])-1
        while l!=0:
            all_data_test.iloc[i+l,-1]+=1
            l=l-1
            
all_data_test=all_data_test[all_data_test["traded"]==0]

all_data_test.drop(columns=["traded"],inplace=True)
```

```python
# There are still some players who did not play during the previous season in the period we are interested in
lst=all_data[all_data["season"]==1].loc[:,"br_id"].unique()

# We can drop them with the exception of Ersan Ilyasova and Hassan Whiteside who have multiple seasons we are interested in
all_data=all_data[~all_data["br_id"].isin(lst[:-2])]

# For Ersan Ilyasova we have to drop his 2011 candidacy
all_data.drop(index=all_data[all_data["br_id"]=="ilyaser01"].iloc[0:2].index,inplace=True)

# For Hassan Whiteside we have to drop his 2015 candidacy
all_data.drop(index=all_data[all_data["br_id"]=="whiteha01"].iloc[0:2].index,inplace=True)
```

```python
# There are some players who haven't played at all in 2022
dnp=all_data_test["br_id"].value_counts()<2
all_data_test=all_data_test[~all_data_test['br_id'].isin(dnp[dnp].index)]
```

```python
# There are 5 players who had minimal playtime during this season or in the previous one and have messed up data
nas=all_data_test.loc[all_data_test["PF"].isna()]["br_id"].unique()
all_data_test=all_data_test[~all_data_test["br_id"].isin(nas)]
```

```python
# Renaming duplicate columns
cols=pd.Series(all_data.columns)
for dup in all_data.columns[all_data.columns.duplicated(keep=False)]: 
    cols[all_data.columns.get_loc(dup)] = ([dup + '.' + str(d_idx) 
                                     if d_idx != 0 
                                     else dup 
                                     for d_idx in range(all_data.columns.get_loc(dup).sum())]
                                    )
all_data.columns=cols

cols=pd.Series(all_data_test.columns)
for dup in all_data_test.columns[all_data_test.columns.duplicated(keep=False)]: 
    cols[all_data_test.columns.get_loc(dup)] = ([dup + '.' + str(d_idx) 
                                     if d_idx != 0 
                                     else dup 
                                     for d_idx in range(all_data_test.columns.get_loc(dup).sum())]
                                    )
all_data_test.columns=cols
```

```python
all_data=all_data.rename({"MP.1": "MP Total"},axis=1)
all_data_test=all_data_test.rename({"MP.1": "MP Total"},axis=1)
```

```python
# Converting columns to numerical values
cols = all_data_test.columns.drop(["br_id","team","league","pos"])

all_data_test[cols]=all_data_test[cols].apply(pd.to_numeric, errors='coerce')
all_data[cols]=all_data[cols].apply(pd.to_numeric, errors='coerce')
```

```python
# Fixing NA values in the PTS column
for i in all_data.index:
    k=len(all_data[all_data["PTS"].isna()].loc[:,"3P%":"PTS"].columns)-1
    if np.isnan(all_data.loc[i,"PTS"]):
        while k!=0:
            all_data.loc[i,all_data[all_data["PTS"].isna()].loc[:,"3P%":"PTS"].columns[k]]=all_data.loc[i,all_data[all_data["PTS"].isna()].loc[:,"3P%":"PTS"].columns[k-1]]
            k=k-1
        all_data.loc[i,"3P%"]=0
        
for i in all_data_test.index:
    k=len(all_data_test[all_data_test["PTS"].isna()].loc[:,"3P%":"PTS"].columns)-1
    if np.isnan(all_data_test.loc[i,"PTS"]):
        while k!=0:
            all_data_test.loc[i,all_data_test[all_data_test["PTS"].isna()].loc[:,"3P%":"PTS"].columns[k]]=all_data_test.loc[i,all_data_test[all_data_test["PTS"].isna()].loc[:,"3P%":"PTS"].columns[k-1]]
            k=k-1
        all_data_test.loc[i,"3P%"]=0
```

```python
# There are still some messed up data in our datasets. Some players have Null Points per game which seems highly unlikely.
# This is because they had null values in 3P % because they didn't have a 3PA per game or FT% because they didn't have
# a FTA per game and the algorithm misplaced the data.

# We want to eliminate those who have null FTs per game value as it is highly unlikely for an MIP candidate to have
# 0 FTAs per game but it's possible for him to have 0 3PA if he is a big man

nas=all_data_test[(all_data_test["PTS"].isna()) & (all_data_test["3PA"]!=0)]["br_id"].unique()
all_data_test=all_data_test[~all_data_test["br_id"].isin(nas)]

nas=all_data[(all_data["PTS"].isna()) & (all_data["3PA"]!=0)]["br_id"].unique()
all_data=all_data[~all_data["br_id"].isin(nas)]
```

```python
# Fixing Nene's 2006 season
idx=all_data[(all_data["br_id"]=="hilarne01") & (all_data["season"]==2006)].index[0]
all_data.at[idx,"BLK"]=0
all_data.at[idx,"TOV"]=2
all_data.at[idx,"PF"]=1
all_data.at[idx,"PTS"]=0
```
