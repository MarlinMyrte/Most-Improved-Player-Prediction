### Creating functions to evaluate and predict

```python
# Create CV function
def cv(x_train_data,y_train_data, test_data, model):
    out = np.zeros(35)
    acc=0
    
    # Calculating accuracy
    for i in range(1,36):
        test_i = fold_vec == i
        train_i = fold_vec != i
        x_test = x_train_data.iloc[test_i, :]
        x_train = x_train_data.iloc[train_i, :]
        y_test= y_train_data[test_i]
        y_train = y_train_data[train_i]
        mod = model
        mod.fit(x_train, y_train)
        prob = mod.predict_proba(x_test)
        prob = prob[:, 1]
        max=prob.max()
        pred=prob==max
        if y_test[pred][0]==1:
            acc+=1
        out[i-1]=accuracy_score(y_test,pred)
        
    print("Average Accuracy: ",np.mean(out))
    #print("Accuracy: ",acc,"/35")
    #print("Accuracy%: ",acc/35)
    
    # Making predictions
    mod=model
    mod.fit(x_train_data, y_train_data)
    mip_prob=mod.predict_proba(test_data)
    mip_prob = mip_prob[:, 1]
    mip = {'br_id': test.index, 'pred': mip_prob}
    mip = pd.DataFrame(data=mip).sort_values("pred",ascending=False).head()
    #mip=mip[mip["pred"]!=1]
    return mip
```

```python
# Create CV function - Oversampling
def cv_OS(x_train_data,y_train_data, test_data, model):
    out = np.zeros(35)
    acc=0
    
    # Calculating accuracy
    for i in range(1,36):
        test_i = fold_vec_OS == i
        train_i = fold_vec_OS != i
        x_test = x_train_data.iloc[test_i, :]
        x_train = x_train_data.iloc[train_i, :]
        y_test= y_train_data[test_i]
        y_train = y_train_data[train_i]
        mod = model
        mod.fit(x_train, y_train)
        prob = mod.predict_proba(x_test)
        prob = prob[:, 1]
        max=prob.max()
        pred=prob==max
        if y_test[pred][0]==1:
            acc+=1
        out[i-1]=accuracy_score(y_test,pred)
        
    print("Average Accuracy: ",np.mean(out))
    #print("Accuracy: ",acc,"/35")
    #print("Accuracy%: ",acc/35)
    
    # Making predictions
    mod=model
    mod.fit(x_train_data, y_train_data)
    mip_prob=mod.predict_proba(test_data)
    mip_prob = mip_prob[:, 1]
    mip = {'br_id': test.index, 'pred': mip_prob}
    mip = pd.DataFrame(data=mip).sort_values("pred",ascending=False).head()
    #mip=mip[mip["pred"]!=1]
    return mip
```

```python
# Create CV function - Regression
def cv_reg(x_train_data,y_train_data, test_data, model):
    out = np.zeros(35)
    acc=0
    
    # Calculating accuracy
    for i in range(1,36):
        test_i = fold_vec == i
        train_i = fold_vec != i
        x_test = x_train_data.iloc[test_i, :]
        x_train = x_train_data.iloc[train_i, :]
        y_test= y_train_data[test_i]
        y_train = y_train_data[train_i]
        mod = model
        mod.fit(x_train, y_train)
        prob = mod.predict(x_test)
        max=prob.max()
        pred=prob==max
        #print(y_test)
        if y_test[pred].index[0]==y_test.sort_values(ascending=False).index[0]:
            acc+=1
            #print("GOT HIM!")
            #print(y_test[pred].index[0])
        out[i-1]=mean_absolute_error(y_test,prob)
        
    print("Average MAE: ",np.mean(out))
    #print("Accuracy: ",acc,"/35")
    #print("Accuracy%: ",acc/35)
    
    # Making predictions
    mod=model
    mod.fit(x_train_data, y_train_data)
    mip_prob=mod.predict(test_data)
    mip = {'br_id': test.index, 'pred': mip_prob}
    mip = pd.DataFrame(data=mip).sort_values("pred",ascending=False).head()
    return mip
```
